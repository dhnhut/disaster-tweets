```python
def load_dataset(
    dataset: str,
    file: str,
    headers: list[str] = [],
    sep: str = ",",
    encoding: str = "utf-8",
    print_head: bool = True,
    map_columns: dict[str, str] = {},
    frac: float = 1.0,
) -> pd.DataFrame:
    filepath = dataset_path / dataset / file
    df = pd.read_csv(
        filepath,
        sep=sep,
        encoding=encoding,
        names=headers if headers else None,
        on_bad_lines="skip",
    )
    if print_head:
        print(f"Shape {df.shape}")
        print(f"Original:")
        display(df.head())

    # Rename columns if mapping is provided and select only the mapped columns
    if map_columns:
        df = df.rename(columns=map_columns)[list(map_columns.values())]

    # Sample the dataset if frac < 1.0
    if frac < 1.0:
        df = df.sample(frac=frac, random_state=setup.RANDOM_SEED)
        print(f"Fraction Shape {df.shape}")

    df["dataset"] = dataset

    if print_head:
        display(df.head())
    return df
```

### A.1. Sanders Analytics Sentiment-labelled tweets

```python
df_1 = load_dataset(
    dataset="Sanders Analytics Sentiment-labelled tweets",
    file="full-corpus.csv",
    map_columns={"TweetText": "tweet_text"},
)

# The dataset contains tweets about 4 companies: Microsoft, Google, Twitter, and Apple.
# df_1['Topic'].value_counts()
# microsoft    1364
# google       1317
# twitter      1290
# apple        1142

companies = ["microsoft", "google", "twitter", "apple"]
for company in companies:
    # remove @company from the text
    df_1["tweet_text"] = df_1["tweet_text"].str.replace(
        f"@{company}", "@USER", case=False
    )
df_1.head()
```

```python
# FULL DATASET
dataset_4 = "Customer Support on Twitter"
file_4 = "twcs/twcs.csv"

# # SAMPLE DATASET
# dataset_4 = 'Customer Support on Twitter'
# file_4 = 'sample.csv'

df_4 = load_dataset(
    dataset=dataset_4,
    file=file_4,
    map_columns={"text": "tweet_text", "author_id": "author_id"},
    frac=0.1,  # sample 10% of the dataset to speed up processing
)

# replace @author_id with @USER in the text
df_4["tweet_text"] = df_4.apply(
    lambda row: row["tweet_text"].replace(f"@{row['author_id']}", "@USER"), axis=1
)
df_4.drop(columns=["author_id"], inplace=True)
df_4.head()
```
