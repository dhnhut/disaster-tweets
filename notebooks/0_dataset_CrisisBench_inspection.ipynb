# CrisisBench - Inspection

Inspect the dataset, whether the train/test/dev set has dataleakage

```python
from pathlib import Path
import os
import csv
import numpy as np
import pandas as pd

import configuration
from src import dataset_settings

from dotenv import load_dotenv

load_dotenv()

dataset_path = Path(os.getenv("DATASETS_PATH")) / "CrisisBench" / "data"
```

#### B.1. all_data_en

```python
base_dir = dataset_path / "all_data_en"

train_path = base_dir / "crisis_consolidated_humanitarian_filtered_lang_en_train.tsv"
dev_path = base_dir / "crisis_consolidated_humanitarian_filtered_lang_en_dev.tsv"
test_path = base_dir / "crisis_consolidated_humanitarian_filtered_lang_en_test.tsv"

df_train = pd.read_csv(
    train_path,
    sep="\t",
    quoting=csv.QUOTE_NONE,
    on_bad_lines="skip",
)
df_dev = pd.read_csv(
    dev_path,
    sep="\t",
    quoting=csv.QUOTE_NONE,
    on_bad_lines="skip",
)
df_test = pd.read_csv(
    test_path,
    sep="\t",
    quoting=csv.QUOTE_NONE,
    on_bad_lines="skip",
)
```

```python
set(df_train["event"].unique()) == set(df_dev["event"].unique()) == set(df_test["event"].unique())
```
