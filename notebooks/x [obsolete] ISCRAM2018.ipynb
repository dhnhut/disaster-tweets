# ISCRAM2018

The dataset contains all tweet id, without labeling, therefore, cannot be used.

---

## A. Original description

This corpora comprised of Twitter data collected during three natural disasters namely Hurricane Harvey, Irma, and Maria in 2017. 

Due to Twitter restrictions, we only share tweet ids. However, we also provide a tool to download full tweets content using the tweet ids.

Data format of most of the files is self-explanatory and is organized as follows:

Hurricane Irma:
- "Irma_tweet_ids.txt" contains 1,207,272 tweet ids.
- "Irma_image_urls.csv" contains 60,974 tweet ids and their corresponding image URLs.

Hurricane Maria:
- "Maria_tweet_ids.txt" contains 1,096,335 tweet ids.
- "Maria_image_urls.csv" contains 19,682 tweet ids and their corresponding image URLs.

Hurricane Harvey:
- "Harvey_tweet_ids.txt" contains 6,506,303 tweet ids.
- "Harvey_image_urls.csv" contains 115,526 tweet ids and their corresponding image URLs.
- "Harvey_missing_ids_50k.csv" contains around 50k tweets content. 

Due to some issues during the data collection, we lost tweet ids of around 225k tweets. That's the reason, out of these tweets, as per Twitter terms, we provide full content of around 50k in the above file. If you want to get the rest of the data (around 175k), then please send us an email.

### Ref

Firoj Alam, Ferda Ofli, Muhammad Imran, Michael Aupetit. A Twitter Tale of Three Hurricanes: Harvey, Irma, and Maria. In Proceedings of the 15th International Conference on Information Systems for Crisis Response and Management (ISCRAM), May 2018, Rochester NY, USA.
