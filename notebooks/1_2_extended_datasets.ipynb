```python
from pathlib import Path
import os

import numpy as np
import pandas as pd

import configuration
from src import dataset_settings, config

from dotenv import load_dotenv
load_dotenv()

dataset_path = Path(os.getenv("DATASETS_PATH")) / 'Datasets for Balancing'
```

```python
def load_dataset(
    filepath: Path,
    map_columns: dict[str, str] = {},
    sep: str = ",",
    encoding: str = "utf-8",
    print_head: bool = True,
    headers: list[str] = [],
    frac: float = 1.0,
) -> pd.DataFrame:
    df = pd.read_csv(filepath, sep=sep, encoding=encoding, names=headers if headers else None, on_bad_lines='skip')
    print(f"Shape {df.shape}")
    if map_columns:
        df = df.rename(columns=map_columns)[list(map_columns.values())]
    if print_head:
        display(df.head())
    if frac < 1.0:
        df = df.sample(frac=frac, random_state=config.RANDOM_SEED)
        print(f"Fraction Shape {df.shape}")
    return df
```
